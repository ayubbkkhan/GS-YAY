# Load the data

```python
df=pd.read_csv('E:/YAY/DailyDelhiClimateTrain.csv')
df.head()
```

### Check the features/columns of the dataset

```python
df.columns
```

### Drop the unneccessary columns / features

```python
to_drop=['meantemp', 'wind_speed', 'meanpressure']
X=df.drop(to_drop,axis=1)         
```

### Group the data. Reset the index with time i.e. 'date'

```python
X=X.groupby('date')['humidity'].sum().reset_index()
X = X.set_index('date')
X.index
```

#### Check what X looks like

```python
X[0:5]
```

#### Now we can plot the time series graph

```python
X.plot(figsize=(15,6))
```

#####  We also can examine for a specified period , say one year

```python
X['2013':'2014'].plot(figsize=(15,6))
```

#### We can look by taking out the series 'humidity' from X 

```python

y=X[['humidity']]
y[0:5]
```

It gives the same output and let's see the graphs also

```python
y['2013':'2014'].plot(figsize=(15,6))
```
